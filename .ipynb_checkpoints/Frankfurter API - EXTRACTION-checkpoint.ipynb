```python
import pandas as pd
import numpy as np

import time
import datetime

import requests

from pprint import pprint
```

## Variables Globales

Variables necesarias para la extracción y organización de la información que nos brinda la API de Frankfurter.

```python
url = "https://api.frankfurter.app"
from_currency = "EUR"
init_date = "1999-01-04"
fin_date = datetime.datetime.today()
amount = 1
```

## Funciones

Función para extraer el glosario de monedas que Frankfurter registra históricamente con una moneda dada.

```python
def get_currencies(from_currency, to_currency, init_date, fin_date, amount):
        
    currencies = list()
    coins = set()
    
    days_dif = datetime.datetime.strptime(datetime.datetime.strftime(fin_date, "%Y-%m-%d"), "%Y-%m-%d") - datetime.datetime.strptime(init_date, "%Y-%m-%d")

    for days in range(days_dif.days + 1):
        
        date = datetime.datetime.strptime(init_date, "%Y-%m-%d") + datetime.timedelta(days = days)
        
        date = datetime.datetime.strftime(date, "%Y-%m-%d")

        endpoint = f"{url}/{date}?amount={amount}&from={from_currency}"

        currencies.append(list(requests.get(endpoint).json()["rates"].keys()))
        
        print(f"{date} completed")
        
    print("-------------First loop completed-------------")
        
    for enum, currencies_x_days in enumerate(currencies):
            
        for currency in currencies_x_days:

            coins.add(currency)
            
        print(f"{enum} list completed")
            
    return sorted(coins)
```

Función para obtener el cambio histórico entre el Euro y el resto de monedas registradas por Frankfurter.

```python
def historical_convertion(currencies, from_currency, init_date, fin_date, amount):
        
    dataframe = pd.DataFrame(columns = currencies)
    
    original_coin_amount = list()
    
    dates = list()
    
    days_dif = datetime.datetime.strptime(datetime.datetime.strftime(fin_date, "%Y-%m-%d"), "%Y-%m-%d") - datetime.datetime.strptime(init_date, "%Y-%m-%d")
        
    for days in range(days_dif.days + 1):
        
        converted_values = dict()
        
        date = datetime.datetime.strptime(init_date, "%Y-%m-%d") + datetime.timedelta(days = days)
        
        date = datetime.datetime.strftime(date, "%Y-%m-%d")

        endpoint = f"{url}/{date}?amount={amount}&from={from_currency}"

        response = requests.get(endpoint).json()
        
        for coin in currencies:
            
            try:
            
                converted_values[coin] = response["rates"][coin]
                
            except:
                
                converted_values[coin] = np.nan
                        
        # Convertir el diccionario en un DataFrame
        converted_values_df = pd.DataFrame.from_dict(converted_values, orient = 'index', columns = [days])

        # Concatenar el nuevo DataFrame al existente (axis=0 para apilar en filas)
        dataframe = pd.concat([dataframe, converted_values_df.T], axis = 0, sort = False)
        
        original_coin_amount.append(amount)
        dates.append(date)
            
        print(f"-----------{date} completed-----------")
        
    dataframe[from_currency] = original_coin_amount
    dataframe["Dates"] = dates
    
    return dataframe
```

## Extracción de datos

```python
%%time

currencies = get_currencies(from_currency, to_currency, init_date, fin_date, amount)
```

```python
print(currencies)
```

```python
%%time

historical = historical_convertion(currencies, from_currency, init_date, fin_date, amount)
```
